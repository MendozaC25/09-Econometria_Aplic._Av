3. [5 points ] Construct a variable for the “treatment” called anykids (indicator for 1 or more kids) and a variable for time being after the expansion (called post93—should be 1 for 1994 and later).

```python
#Se asignará el valor de 1 para indicar que tiene hijos, y 0 en caso no los tenga

def treatment(var):
    if var >= 1:
        return 1
    else:
        return 0
```

```python
eitc['anykids'] = eitc['children'].apply(treatment)
```

```python
# Comprobamos que se ha creado correctamente la variable

eitc.iloc[:,[3,11]]
```

```python
#Se asignará el valor de 1 para indicar que es posterior a 1993, y 0 en caso no lo sea

def post(var):
    if var > 1993:
        return 1
    else:
        return 0
```

```python
eitc['post93'] = eitc['year'].apply(post)
```

```python
# Comprobamos que se ha creado correctamente la variable

eitc.iloc[148:158,[1,12]]
```

4. [10 points ] Using the “interaction term” diff-in-diff specification, run a regression to estimate the difference-in-differences estimate of the effect of the EITC program on earnings. Use all women with children as the treatment group.

```python
# Creamos el término de interacción

eitc['int_term'] = eitc['anykids'] * eitc['post93']
```

```python
# Corremos la regresión Diff in Diff

import statsmodels.formula.api as smf

modelo1 = smf.ols('earn ~ anykids + post93 + int_term', data=eitc).fit()
print(modelo1.summary())
```
